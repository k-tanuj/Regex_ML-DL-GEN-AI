### Encoding - transform categorical data into numerical data

1. Label Encoder -> 1-1 column encode
2. Ordinal Encoder -> we can encode more than 2 cols with respect to their sub categories.
3. OneHotEncoder -> create new column based on the sub categories. it doesnt have col name, shows only values without col names.
4. Get Dummies (Pandas) -> Create new col based on sub categories but it has col name also

```python
import numpy as np 
import pandas as pd
```

### 2. Ordinal Encoder


```python
df = pd.read_csv('day 6 data\\covid_toy.csv')
df.head(2)
```

```python
df = df.drop(columns=["age","fever",'city'])
```

```python
df.head(2)
```

```python
df['gender'].value_counts()
```

```python
df['has_covid'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Male','Female'],
                    ['Mild','Strong'],
                    ['Yes','No']])
```

```python
oe
```

```python
df_oe = oe.fit_transform(df) ## fit means to learn the categories and transform means to apply the encoding
```

```python
df_new = pd.DataFrame(df_oe, columns = df.columns)
```

```python
df_new.head()
```

### OneHot Encoder
    ->has_covid(no,yes)-> has_covid_no, has_covid_yes

```python
df = pd.read_csv('day 6 data\\covid_toy.csv')
df.head(2)
```

```python
df = df.drop(columns=["age","fever"])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output=False,dtype=np.int32) ## drop = 'first' is used to avoid dummy variable trap and sparse_output=False is used to get the output in dense format
```

```python
df_new= ohe.fit_transform(df[['gender','cough','has_covid','city']])
```

```python
df_new
```

```python
df_new =pd.DataFrame(df_new, columns = ohe.get_feature_names_out())
df_new.head()
```

```python
df['has_covid'].value_counts()
```

### GET DUMMIES

```python
df = pd.read_csv('day 6 data\\covid_toy.csv')
df.head(2)
```

```python
df = df.drop(columns=["age","fever"])
```

```python
df.head(2)
```

```python
df = df.dropna()
```

```python
df_new = pd.get_dummies(df,columns=['gender','cough','has_covid','city'] ,drop_first=True)
```

```python
df_new = pd.DataFrame(df_new,columns = df_new.columns)
```

```python
df_new = df_new.astype(int)
```
