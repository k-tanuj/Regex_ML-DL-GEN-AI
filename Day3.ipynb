# EDA - EXPLORATORY DATA ANALYSIS

### Parts of EDA
    1.Univariate Analysis -> Analysis on a single independednt col.
    2.Bivariate Analysis -> on 2 cols.
    3.Multivariate Analysis -> on more than 2 cols.

### Datatypes
    1.Numerical data: continuos data -> that can be categorized (AGE[years,months.days], height,weight)
    2. Categorical Data : discrete data -> Total no. of employess, marital status,etc.
