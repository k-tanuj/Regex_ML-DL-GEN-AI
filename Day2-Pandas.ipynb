# DAY 2 - PANDAS

Pandas is used to handle data manipulations.

```python
import pandas as pd
```

1. Series -> one dimensional array, it shows only values, not column names.

```python
a=pd.Series([1,2,3,4,5])
print(type(a))
a
```

2. DataFrame -> multidimensional array and it shows col name with thier values.

```python
a ={
    "Id": [1,2,3,4,5],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Dept": ["Sales", "IT", "IT", "IT", "Sales"],
    "Salary": [25000, 30000, 35000, 40000, 45000],
    "Working Hours": [8, 9, 7, 8, 6], 
} #dictionary of lists

df= pd.DataFrame(a)
print(type(df))
df
```

Export or import a file

```python
##exporting to csv
#with index
df.to_csv("emp.csv")
#without index
df.to_csv("emp2.csv", index=False)
```

```python
#importing from csv
#give absolute path if the file is not in the same directory
df1=pd.read_csv("emp2.csv")
df1
```
