Workflow in ML Models

### Data Dividation
    df : x(input),y(target)
    x : x_train,x_test
    y : y_train,y_test
    ML MODEL (x_train,y_train)->model(x_test)->o/p ??y_pred
    compare : y_test and y_pred

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('covid_toy.csv')
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df["fever"] = df['fever'].fillna(df['fever'].mean())
```

```python
df.isnull().sum()
```

```python
df.head(2)
```

```python
df.shape
```

```python
x = df.drop(columns = ["has_covid"]) ##input data
y = df['has_covid'] ##output data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2,random_state = 42)
```

```python
print("Total Data Shape: ", df.shape)
print("----------------------------")
print("Input Data Shape: ", x.shape)
print("x_train Data Shape: ", x_train.shape)
print("x_test Data Shape: ", x_test.shape)
print("----------------------------")
print("Output Data Shape: ", y.shape)
print("y_train Data Shape: ", y_train.shape)
print("y_test Data Shape: ", y_test.shape)
```

### Normal Distribution convert -> calculations easy ,less model training time

    Normal Distribution Condition -> mean=0, std_deviation(next value)=1

### ENCODING -> Transforming categorical data into numerical data

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender'] = lb.fit_transform(df['gender'])
df['cough'] = lb.fit_transform(df['cough'])
df['city'] = lb.fit_transform(df['city'])
df['has_covid'] = lb.fit_transform(df['has_covid'])

```

```python
df.head()
```

```python
x= df.drop(columns = ["has_covid"]) ##input data
y = df['has_covid'] ##output data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2,random_state = 42)
```

```python
np.round(x_train.describe(),2 )
```

```python
# for converting our data into normal distribution we will use standard scaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc =sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc,columns=x_train.columns)
```

```python
np.round(x_train_new.describe(),2   )
```
