Functions in Numpy

1. zeros() -> it will create an array in which all elements are zero.

```python
a = np.zeros(3)
a
```

```python
a=np.zeros((2,3))
a
```

2. ones() -> Will create an array in which all elements are one.

```python
a = np.ones((2,3))
a
```

3. eye(): will create an identity matrix

```python
a = np.eye(3,3)
a
```

4. diag(): we can give custom values for diagonal elements.

```python
a = np.array([1,45,67,99])

b = np.diag(a)
b
```

5. randint() -> provides random integers in given range. 
 Syntax:np.random.randint(min_value,max_value,no. of elements)

```python
a= np.random.randint(1,9,3)
a
```

6. rand()-> it will create an array in which random numbers will be in range b/w 0 to 1.

```python
a= np.random.rand(5)
a
```

7. seed() -> it will fix your random generated numbers.

```python
np.random.seed(24)
a=np.random.randint(1,8,3)
a
```

8.  view and copy ->
      #->view : modification in original data.
      #->copy : changes in duplicate data will not reflect in original data.


```python
#view
a = np.array([1, 2, 3,4,3,5,2,5,6,6])
b= a[0:5]
b[:]=0
a 
# values in og dta will be changed as view is created on original data
```

```python
#copy
a = np.array([1, 2, 3,4,3,5,2,5,6,6])
b= a[0:5].copy()
b[:]=0
a
#values in og data will not be changed as copy is created on original data
```

#### Hstack() and vstack():
    1.hstack() : combines 2 or more arrays horizontally
    2.vstack() : combines 2 or more arrays vertically
    

```python
a=np.array([1,2,3])
b=np.arange(4,7)
c=np.arange(7,10)
```

```python
np.hstack((a,b,c)) #horizontal stacking
```

```python
np.vstack((a,b,c)) #vertical stacking 
```
