# Practise Encoding on a insuarance dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("day 6 data\\insurance.csv")
df.head(2)
```

1. Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['sex'] = lb.fit_transform(df['sex'])
df['smoker'] = lb.fit_transform(df['smoker'])
df['region'] = lb.fit_transform(df['region'])

```

```python
df.head()
```

2. Ordinal Encoding

```python
df = pd.read_csv("day 6 data\\insurance.csv")
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = df.drop(columns=["age","bmi",'children','charges'])
```

```python
df.head()
```

```python
df['region'].value_counts()
```

```python
oe = OrdinalEncoder(categories = [['male','female'],
                    ['yes','no'],
                    ['southeast','southwest','northeast','northwest']])
```

```python
df_oe = oe.fit_transform(df) ## fit means to learn the categories and transform means to apply the encoding
```

```python
df_new = pd.DataFrame(df_oe, columns = df.columns)
```

```python
df_new.head()
```

### ONE HOT ENCODING

```python
df = pd.read_csv("day 6 data\\insurance.csv")
df.head(2)
```

```python
df = df.drop(columns=["age","bmi",'children','charges'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output=False,dtype=np.int32) ## drop = 'first' is used to avoid dummy variable trap and sparse_output=False is used to get the output in dense format
```

```python
df_new= ohe.fit_transform(df[['sex','smoker','region']])
```

```python
df_new
```

```python
df_new =pd.DataFrame(df_new, columns = ohe.get_feature_names_out())
df_new.head()
```

4. GET DUMMIES ENCODING

```python
df = pd.read_csv("day 6 data\\insurance.csv")
df.head(2)
```

```python
df = df.drop(columns=["age","bmi",'children','charges'])
```

```python
df.head(2)
```

```python
df_new = pd.get_dummies(df,columns=['sex','smoker','region',] ,drop_first=True)
```

```python
df_new = df_new.astype(int)
```
